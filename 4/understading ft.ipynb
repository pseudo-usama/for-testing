```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-2*np.pi, 2*np.pi, 1000)
```

```python
y1 = np.sin(x) + 3*np.sin(x*2) + np.sin(x)
y2 = y1 - 2*np.sin(x/10)
plt.plot(x, y1)
plt.plot(x, y2)
```

```python
yfft1 = np.fft.fft(y1)
yfft2 = np.fft.fft(y2)
plt.plot(x[:10], np.abs(yfft1)[:10])
plt.plot(x[:10], np.abs(yfft2)[:10])
```

```python
y1_2 = np.abs(np.fft.ifft(yfft1))
y2_2 = np.abs(np.fft.ifft(yfft2))
plt.plot(x, y1_2)
plt.plot(x, y2_2)
```
